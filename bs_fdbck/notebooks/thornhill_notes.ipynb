```python
import pandas as pd
```

```python
table_thornhill = pd.DataFrame()

table_thornhill['models'] = ['UKESM', 'NorESM','GFDL-ESM','CESM','GISS']
table_thornhill['ERF'] = [.03, -.69,-0.45,-.36,-0.24]
table_thornhill['ERF/emissions'] = [.4e-4,-11.8e-4,-9.7e-4,-5.4e-4,-1.3e-4]
table_thornhill['emissions/dT'] = [32,234,81,156,113]
table_thornhill['alpha'] = [0.001,-.28,-.079,-.084,-.015]

table_thornhill = table_thornhill.set_index('models')
```

```python
import matplotlib.pyplot as plt
```

```python
fi, axs = plt.subplots(1,4, figsize=[10,5])
for c,ax in zip(table_thornhill.columns, axs):
    table_thornhill[c].plot.bar(ax=ax)
    ax.set_title(c)
#    plt.show()
plt.tight_layout()
```
