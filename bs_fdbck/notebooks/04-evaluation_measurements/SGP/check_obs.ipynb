```python
path_acsm = measurements_path/'ARM_data' /'watson'/'ACSM'
```

```python
path_acsm
```

```python

fl = list(path_acsm.glob('*.csv'))
```

```python
fl.sort()
```

```python
df_list = []
for f in fl:
    df_list.append(pd.read_csv(f, index_col=0))
```

```python
df_acsm = pd.concat(df_list)
```

```python
df_acsm.index = pd.to_datetime(df_acsm.index)
```

```python
df_daily_mean = df_acsm.resample('M').mean()#.plot.area()
```

```python
df_daily_median = df_acsm.resample('M').median()#.plot.area()
```

```python
df_daily_mean[df_daily_mean>0].plot()
```

```python
df_daily_mean[df_daily_mean>=0].plot.area(stacked=True)
```

```python
df_daily_median[df_daily_median>0].plot()
```

```python
df_daily_median[df_daily_median>=0].plot.area(stacked=True)
```
