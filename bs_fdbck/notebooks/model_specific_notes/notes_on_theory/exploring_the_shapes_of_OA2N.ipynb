Let's assume that the distribution is completely uniform with $D_p$. Next assume that we are in the kinetic regime and the growth rate is approximately constant with size ($dD/dt$ is independent of $D$). 

How much mass does it take to move a particle of size $d_0$ over a limit $d_l$? 

Define $\Delta  = d_l-d_0$. 

The difference in mass is

$$
\begin{align}
\Delta M = & \rho\frac{1}{6}\pi \Big(d_l^3-(d_l-\Delta D)^3\Big) \\
 =&  \rho\frac{1}{6}\pi \Big( 3d_l^2\Delta D -3 d_l \Delta D^2 + \Delta D^3 \Big) \\
 =&  \rho\frac{1}{6}\pi \Big(\Delta D^3 -3 d_l \Delta D^2+ 3d_l^2\Delta D    \Big) \\
\end{align}
$$
Meaning the $dm = a(dD)^3-b(dD)^2+ c (dD)$

Now note that if the distribution is uniform, then the number change is directly proportional to $\Delta D$, that is if we only have one step function of particles.



```python
dl = 100
dd = np.linspace(0,30)
dm = 1/6.*np.pi*(dd**3-3*dl*dd**2+3*dl**2*dd)

fig, ax = plt.subplots(dpi=100)
plt.plot(dm, dd )
plt.ylabel('$\Delta D$ [nm] or $\Delta N$')
plt.xlabel('$\Delta$ mass [??]')
```

With reasonable values for dl and dd (in this case dl =100, and dd from 0 to 30), it's pretty close to linear. 

### What does this mean? It means that we do not get a flattening of from no change in size or processes and a constant GR

What if we include some more particles? Say that we grow all the particles from $D$ to $D+ \Delta D$

$$

```python
dl = 100
dd = np.linspace(0,30)
dm = ((dl +dd)**4-dl**4)

fig, ax = plt.subplots(dpi=100)
plt.plot(dm, dd )
plt.ylabel('$\Delta D$ [nm] or $\Delta N$')
plt.xlabel('$\Delta$ mass [no sense units]')
```

### What does this mean? Just by assuming a uniform distribution and growing all of them a constand $\Delta D$ we get a leveling off because of the increase in surface area with growth.

## What are we missing in this model? 

- We are missing that the distribution will not look like this. Likely we will e.g. have more particles below $d_l$ instead of the same. Or in principle could be less, but unlikely to be completely flat. 
- We are missing that GR is likely dependent on number and surface area. When there is a lot of particles, the mass will be distributed to more particles. This is tricky to take into account in a simplified model. 

#### How would it look to take into accound a change in sizedist? 



Assume that $\frac{d^2N}{d^2D}=-a$
This means that in the simplified model above we will assume we have N particles but the distribution is $\frac{d^2N}{d^2D}=-a$
This means the  $\frac{dN}{dD} \propto N(1-aD)$



$$
\begin{align}
\rho \frac{1}{6} \pi \int_{d_l-\Delta D}^{d_l} N(1-aD)((D+\Delta D)^3 -D^3)dD =& \rho N \frac{1}{6} \pi \Big[ \frac{1}{4} (D+\Delta D)^4 - \frac{1}{4}D^4 +C \Big]_{d_l-\Delta D}^{d_l} - a \rho N \frac{1}{6} \pi \int_{d_l-\Delta D}^{d_l} D(D+\Delta D)^3 -D^4) dD \\
 = &  \rho \frac{1}{24} \pi \Big[  2(d_l+\Delta D)^4 - 2d_l^4  \Big]- a \rho N \frac{1}{6} \pi \int_{d_l-\Delta D}^{d_l} D(D+\Delta D)^3 -D^4) dD \\
=& \Gamma - a \rho N \frac{1}{6} \pi \Big[   \frac{3}{4}(\Delta D) D^4  +(\Delta D) ^2 D^3 + \frac{(\Delta D)^3}{2}D^2\Big]_{d_l-\Delta D}^{d_l}  \\
=& \rho \frac{1}{24} \pi \Big[  2(d_l+\Delta D)^4 - 2d_l^4  -  a \Big(\frac{3}{4}\Delta D (d_l^4-(d_l-\Delta D)^4) + (\Delta D)^2 (d_l^3-(d_l-\Delta D)^3)+ \frac{(\Delta D)^3}{2} (d_l^2-(d_l-\Delta D)^2) \Big) \Big] \\
\end{align}
$$

the function N(1-aD) above doesn't reflect the true alpha, must be normalized by some D in order to be correct. 

## For the same GR you would expect a stronger increase for smaller limits because there are so many more. 

```python
dl = 100
dd = np.linspace(0,30)
fig, ax = plt.subplots(dpi=100)

for v in np.linspace(0,1,5):
    a = v/dl

    dm = 1/4*dd**2*(     a*dd**3   +   dd**2*(2-4*a*dl)  +  6*a*dd*dl**2  -   12*dl**2*(a*dl-1))
    plt.plot(dm, dd , label = f'a={v}')
plt.legend()
plt.ylabel('$\Delta D$ [nm] ')
plt.xlabel('$\Delta$ mass [no sense units]')
```

## But this does only look at $\Delta D$. But if we have $dN/dD = (1-aD)$ in the $dN = D(1-0.5aD)$

```python
dl = 100
dd = np.linspace(0,30)
fig, ax = plt.subplots(dpi=100)

for v in np.linspace(0,1,5):
    a = v/dl

    dm = 1/4*dd**2*(     a*dd**3   +   dd**2*(2-4*a*dl)  +  6*a*dd*dl**2  -   12*dl**2*(a*dl-1))
    dN = dd*(1-a/2*dd)
    plt.plot(dm, dN , label = f'a={v}')
plt.legend()
plt.ylabel('$\Delta N$ [#] ')
plt.xlabel('$\Delta$ mass [no sense units]')
```
