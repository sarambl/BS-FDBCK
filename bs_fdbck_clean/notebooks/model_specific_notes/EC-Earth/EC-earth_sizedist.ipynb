```python
def add_diameter_mode(ds):
    for r in rad_vars:
        if r in ds:
            d = f'D{r[1:]}'
            ds[d] = ds[r]*2            
            ds[d].attrs['units'] = ds[r].attrs['units']
            #ds[d].attrs['long_name'] = ds[r].attrs['units']
    return ds

ds_sub = add_diameter_mode(ds_sub)
```

Taking the geometric standard deviation from documentation: 
- $S=2$ for coarse mode
- $S=1.59$ for all the rest

Then, the model outputs geometric mean/aritmetic median radius of the mode, $r_m$, and we calculate the geometric mean diameter, $d_m$ by multiplying with two.
The total number in the mode is denoted by $N_0$
Finally we use:

\begin{equation}
\frac{dN}{dlnD} = \frac{N_0}{\sqrt{2\pi}}\frac{1}{ln(S)} \exp \Big[ - \frac{(ln(d)-ln(d_m))^2}{2ln^2(S)} \Big]
\end{equation}
To get in log10, we use $\frac{dln(D)}{dD}=\frac{1}{D}$ and $\frac{dlog10(D)}{dD}=\frac{1}{Dln(10)}$, which means that: 
\begin{align}
\frac{dN}{dlog10D} = &  \frac{dN}{dlnD} \cdot \frac{dln(D)}{dD} \cdot \frac{dD}{dlog10(D)} \\
\frac{dN}{dlog10D} = &  \frac{dN}{dlnD} \cdot \frac{1}{D} \cdot \frac{Dln(10)}{1} \\
\frac{dN}{dlog10D} = &  \frac{dN}{dlnD} \cdot ln(10) \\
\frac{dN}{dlog10D} \approx &  2.3 \cdot \frac{dN}{dlnD} \\
\end{align}

```python
for mo in modes_dic:
    _di = modes_dic[mo]
    vnum = _di['NUM']
    vrad = _di['RAD']
    vdiam = _di['DIAM']
    sig = _di['SIG']
    
    num = ds_sub[vnum]
    rad = ds_sub[vrad]
    diam = ds_sub[vdiam]
    
    radiuses = ds_pnsd['diameter']/2
    diameters = ds_pnsd['diameter']
    
    exponent = (np.log(diameters)-np.log(diam))**2/(2*np.log(sig)**2)
    
    first_term = num/(np.sqrt(2*np.pi)*np.log(sig))
    dNdlogD = first_term*np.exp(-exponent)
    ds_pnsd[f'dNdlogD_{mo}'] = dNdlogD
```

```python
from scipy.stats import lognorm
def fix_units_ec_earth(ds):
    for v in num_vars:
        if v in ds:
            if ds[v].attrs['units'] =='1 m-3':
                print(f'Converting {v} from m-3 to cm-3')
                with xr.set_options(keep_attrs=True):
                    ds[v] = ds[v]*1e-6
                ds[v].attrs['units'] = 'cm-3'
    for v in rad_vars:
        if v in ds:
            if ds[v].attrs['units'] =='m':
                print(f'Converting {v} from m to nm')
                with xr.set_options(keep_attrs=True):
                    ds[v] = ds[v]*1e9
                ds[v].attrs['units'] = 'nm'
    for v in diam_vars:
        if v in ds:
            if ds[v].attrs['units'] =='m':
                print(f'Converting {v} from m to nm')
                with xr.set_options(keep_attrs=True):
                    ds[v] = ds[v]*1e9
                ds[v].attrs['units'] = 'nm'

    ds = add_diameter_mode(ds)
    return ds
fix_units_ec_earth(ds_sub)
```

```python
def get_Nx_ec_earth_mod(ds,mod, fromx=100,tox=100000):
    _di = modes_dic[mod]
    vnum = _di['NUM']
    vrad = _di['RAD']
    vdiam = _di['DIAM']
    sig = _di['SIG']
    num = ds[vnum]
    rad = ds[vrad]
    diam = ds[vdiam]

    Nx = num * (lognorm.cdf(tox, np.log(sig), scale=diam)) - num * (
                lognorm.cdf(fromx, np.log(sig), scale=diam))
    return Nx

def get_Nx_ec_earth(ds, fromx=100,tox=100000):
    Nx = 0
    for mod in modes_dic.keys():
        Nxmod = get_Nx_ec_earth_mod(ds, mod, fromx=fromx, tox=tox)
        Nx +=Nxmod
    return Nx

def add_Nx_to_dataset(ds, x_list =[50,70,100,150,200,500],xunit ='nm', add_to_500=True):
    ds = fix_units_ec_earth(ds)
    
    for x in x_list:
        ds[f'N{x}'] = get_Nx_ec_earth(ds,fromx=x)
    if add_to_500 and (500 in x_list):
        for x in x_list:
            if x==500: continue
            nva = f'N{x}'
            nva2 = f'N{x}-500'
            ds[nva2] = ds[nva]-ds['N500']
    return ds
            
    

    
ds_sub = add_Nx_to_dataset(ds_sub)
```
